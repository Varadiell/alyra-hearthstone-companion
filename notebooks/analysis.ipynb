```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/sarahcrowder/8-years-of-hearthstone-decks/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()
```

```python
# Infos
record_set_df.info()
```

```python
# Change columns names
record_set_df.columns = [
    "title",
    "class",
    "expansion",
    "minion_count",
    "spell_count",
    "dust_cost",
    "deck_type",
    "deck_archetype",
    "last_updated",
    "creator",
    "deck_code",
    "rating",
    "cards",
    "card_quantities",
    "card_costs",
    "link"
]

record_set_df.info()
```

```python
# Remove unused columns
record_set_df = record_set_df[["title", "class", "expansion", "deck_type", "deck_archetype", "cards", "card_quantities"]]

record_set_df.info()
```

```python
import ast

# Decoding binary strings into regular strings for all relevant columns
for column in ["title", "class", "expansion", "deck_type", "deck_archetype", "cards", "card_quantities"]:
    # Check if column data contains binary strings and decode
    if record_set_df[column].dtype == object: # Ensure it's an object type column
        record_set_df[column] = record_set_df[column].apply(lambda x: x.decode("utf-8") if isinstance(x, bytes) else x)
        # Convert string representations of lists into actual lists (for 'cards' and 'card_quantities')
        if column in ["cards", "card_quantities"]:
            record_set_df[column] = record_set_df[column].apply(
                lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith("[") else x
            )

record_set_df.head() # Display the first few rows after decoding
```

```python
# Keep only deck_type "Ranked Deck"
ranked_decks_df = record_set_df[record_set_df["deck_type"] == "Ranked Deck"]

ranked_decks_df.info()
```

```python
# Remove decks that have no archetype ("Unknown")
ranked_decks_with_archetype_df = ranked_decks_df[ranked_decks_df["deck_archetype"] != "Unknown"]

ranked_decks_with_archetype_df.info()
```

```python
# Remove "demonhunter", "dream" and "neutral" classes
filtered_df = ranked_decks_with_archetype_df[~ranked_decks_with_archetype_df['class'].str.lower().isin(['demonhunter', 'neutral', 'dream'])]

filtered_df.info()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each class
class_counts = filtered_df['class'].value_counts()

# Custom color list
custom_colors = ['#FFFFFF', '#3FC7EB', '#FFF468', '#8788EE', '#AAD372', '#F48CBA', '#FF7C0A', '#0070DD', '#C69B6D']

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=custom_colors)
plt.title('Repartition of Classes')
plt.show()
```

```python
# Count the occurrences of each expansion
expansion_counts = filtered_df['expansion'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(expansion_counts, labels=expansion_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Repartition of Expansions')
plt.show()
```

```python
# Count the occurrences of each deck archetype
deck_archetype_counts = filtered_df['deck_archetype'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(deck_archetype_counts, labels=deck_archetype_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Repartition of Deck Archetypes')
plt.show()
```

```python
# Repartition of Priest Deck Archetypes
priest_decks = filtered_df[filtered_df["class"] == "priest"]
deck_archetype_counts = priest_decks['deck_archetype'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(deck_archetype_counts, labels=deck_archetype_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Repartition of Priest Deck Archetypes')
plt.show()
```

```python
# Number of Priest decks
priest_decks.info()
```

```python
# Keep the 13 most popular priest deck archetypes ("Control Priest", "Dragon Priest", "N'Zoth Priest", "OTK Priest", "Reno Priest", "Combo Priest", "Silence Priest", "Resurrect Priest", "C'thun Priest", "Highlander Priest", "Quest Priest", "Big Priest" and "Spiteful Priest")
most_popular_priest_deck_archetypes_df = priest_decks[priest_decks['deck_archetype'].str.lower().isin(['control priest', 'dragon priest', 'n\'zoth priest', 'otk priest', 'reno priest', 'combo priest', 'silence priest', 'resurrect priest', 'c\'thun priest', 'highlander priest', 'quest priest', 'big priest', 'spiteful priest'])]

most_popular_priest_deck_archetypes_df.info()
```

```python
# Repartition of the 13 most popular Priest Deck Archetypes
deck_archetype_counts = most_popular_priest_deck_archetypes_df['deck_archetype'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(deck_archetype_counts, labels=deck_archetype_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Repartition of the 13 most popular Priest Deck Archetypes')
plt.show()
```

```python
# Remove unused columns
priest_df = most_popular_priest_deck_archetypes_df[["deck_archetype", "cards", "card_quantities"]]

priest_df.info()
```

```python
priest_df.head()
```

```python
# Convert card-quantity pairs into a dictionary (deck-level)
def count_vectorize(cards, quantities):
    # Clean card names by stripping leading spaces and trailing \r
    cleaned_cards = [card.strip() for card in cards]
    return dict(zip(cleaned_cards, quantities))  # Maintain full card names

# Apply count vectorization across rows and normalize the results
vectorized_df = pd.json_normalize(
    priest_df.apply(lambda row: count_vectorize(row['cards'], row['card_quantities']), axis=1)
).fillna(0).astype('int')

vectorized_df.head()
```

```python
# Remove leading spaces from column names
vectorized_df.columns = vectorized_df.columns.str.lstrip()
vectorized_df.head()
```

```python
# Reset index to align DataFrames
priest_df_reset = priest_df.reset_index(drop=True)
vectorized_df_reset = vectorized_df.reset_index(drop=True)

merged_df = pd.concat([priest_df_reset, vectorized_df_reset], axis=1)
merged_df.head()
```

```python
# Remove unused "cards" and "card_quantities" columns
merged_df.drop(['cards'], axis=1, inplace=True)
merged_df.drop(['card_quantities'], axis=1, inplace=True)
merged_df.head()
```

```python
# Export to csv
merged_df.to_csv('../data/priest_popular_archetype_decks.csv', index=False)
```
