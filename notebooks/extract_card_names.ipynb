```python
# Import the data

import pandas as pd

df = pd.read_csv('../data/priest_popular_archetype_decks.csv')
df.info()
```

```python
df.head()
```

```python
# Remove "deck_archetype" column

df.drop(['deck_archetype'], axis=1, inplace=True)
df.head()
```

```python
# Get the card names of each column

card_names = df.columns.tolist()
print(card_names)
```

```python
# Create a new DataFrame with a single column "card_name"

card_names_df = pd.DataFrame(card_names, columns=['card_name'])

card_names_df.head()
```

```python
# Export to csv with values sorted alphabetically
card_names_df.sort_values(by=['card_name'], ignore_index=True).to_csv('../data/card_names.csv', index=False)
```
